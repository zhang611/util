# 一维高斯分布


$$
f (x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 一维高斯分布参数
mu = 0  # 均值
sigma = 1  # 标准差

# 生成数据
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)  # 在均值附近取值范围
y = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# 绘制一维高斯分布
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f"1D Gaussian: $\\mu={mu}, \\sigma={sigma}$", color='blue')
plt.title("1D Gaussian Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

# 二维高斯分布
$$
f (\mathbf{x}) = \frac{1}{(2\pi)^{d/2} |\mathbf{\Sigma}|^{1/2}} e^{-\frac{1}{2} (\mathbf{x} - \mathbf{\mu})^\top \mathbf{\Sigma}^{-1} (\mathbf{x} - \mathbf{\mu})}
$$

```python

# 二维高斯分布参数
mu_2d = np.array([0, 0])  # 均值向量
sigma_2d = np.array([[1, 0.5], [0.5, 1]])  # 协方差矩阵

# 生成二维网格数据
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# 二维高斯分布的密度函数
def multivariate_gaussian(pos, mu, sigma):
    n = mu.shape[0]
    sigma_det = np.linalg.det(sigma)
    sigma_inv = np.linalg.inv(sigma)
    norm_const = 1 / (np.sqrt((2 * np.pi)**n * sigma_det))
    diff = pos - mu
    return norm_const * np.exp(-0.5 * np.sum(diff @ sigma_inv * diff, axis=-1))

Z = multivariate_gaussian(pos, mu_2d, sigma_2d)

# 绘制二维高斯分布等高线
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar(label='Probability Density')
plt.title("2D Gaussian Distribution")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(alpha=0.3)
plt.show()
```
