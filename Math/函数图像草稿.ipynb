```python
import numpy as np
import matplotlib.pyplot as plt

#  x  0.1  10 log(0) 
x = np.linspace(0.1, 10, 400)

#  y 
y_log = np.log(x) * np.log(x)
y_linear = x

# 
plt.figure(figsize=(10, 6))
plt.plot(x, y_log, label='y = log(x) * log(x)', color='blue')
plt.plot(x, y_linear, label='y = x', color='red')
plt.title('Plot of y = log(x) * log(x) and y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.ylim(-1, 10)  #  y 
plt.legend()
plt.show()
```
