# 圆内随机生点

随机半径,随机角度

半径要用sqrt(rand())

```python
import numpy as np
import matplotlib.pyplot as plt

def random_point_in_circle(R, num_points=1000):
    """ 在圆内均匀生成随机点 """
    theta = np.random.uniform(0, 2 * np.pi, num_points)  # 随机角度
    r = np.sqrt(np.random.uniform(0, 1, num_points)) * R  # 半径均匀分布
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    return x, y

# 生成 1000 个点
R = 1.0  # 圆的半径
x, y = random_point_in_circle(R, num_points=1000)

# 画出点
plt.figure(figsize=(6,6))
plt.scatter(x, y, s=5, color='b', alpha=0.6)  # 画点
plt.xlim(-R, R)
plt.ylim(-R, R)

# 画出圆的边界
circle = plt.Circle((0, 0), R, color='r', fill=False, linewidth=2, linestyle='dashed')
plt.gca().add_patch(circle)

plt.gca().set_aspect('equal')  # 保持XY轴等比
plt.title("Random Points Inside a Circle")
plt.show()
```
